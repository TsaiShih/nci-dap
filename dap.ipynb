```python
#import xml.etree.ElementTree as ET
import xml.etree.ElementTree as ET
# build import_xml function
def import_xml(xml):
    #adding exception
    try:
        tree = ET.parse(xml)
        xml_root = tree.getroot()
        return xml_root
    except FileNotFoundError:
    #file open error
        print('Can not find the file {}'.format(xml))
    

xml_data = import_xml('NorthWind.xml')
```

```python
# ET.tostring function print it out have look
print(ET.tostring(xml_data, encoding='utf8').decode('utf8'))
```

```python
# Using the function I created
xml_data = import_xml('NorthWind.xml')

def find_product(data):
    # Using for loop to find product area
    for i in range(0,len(data)):
        # if the tag show product then do next for loop
        if data[i].tag == "Products":
            # Look into product section to find ProductName and UnitPrice for 2 4 6 8 10 products
            for j in range(0,10,2):
                find_PName = data[i][j].find('ProductName').text
                find_Uprice = data[i][j].find('UnitPrice').text
                print("Product name: {}  Unit price: {}".format(find_PName, find_Uprice))
        

find_product(xml_data)    
```
